```python

# Phase 2 report figure export

# Inference cost comparison

import matplotlib.pyplot as plt

# Final Phase 2 speed checked inference results
models = ["Image Only", "Image + Text"]
ms_per_sample = [3.95, 6.10]

# Creating a bar chart
fig, ax = plt.subplots(figsize=(6, 4))

bars = ax.bar(
    models,
    ms_per_sample,
    edgecolor="black",
    linewidth=1.2
)

# Hatch patterns are added so that bars remain distinguishable even if the report is printed in black and white.
hatches = ["//", "xx"]

for bar, hatch in zip(bars, hatches):
    bar.set_hatch(hatch)

ax.set_ylabel("Milliseconds per Sample")
ax.set_title("Inference Cost Comparison")
ax.set_ylim(0, 7)

# Adding exact values above bars
for i, value in enumerate(ms_per_sample):
    ax.text(
        i,
        value + 0.12,
        f"{value:.2f}",
        ha="center",
        va="bottom"
    )

# Removing unnecessary borders for a cleaner report figure
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

plt.tight_layout()

# Exporting at 300 dpi according to the report guideline
output_file = "phase2_inference_cost_comparison_report.png"
plt.savefig(output_file, dpi=300, bbox_inches="tight")
plt.show()

print(f"Figure saved as: {output_file}")
```
